```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import precision_score, recall_score




from sklearn.model_selection import train_test_split
import os
import glob
from tensorflow import keras
from keras.models import Model
from tensorflow.keras.layers import Input
from keras import Sequential
from keras.layers import Dense, Input
from keras.layers import Permute, Reshape, Conv2D, BatchNormalization, MaxPooling2D, Dropout, Activation, Bidirectional, GRU, TimeDistributed, Dense, Conv1D, LSTM,  Flatten
import matplotlib.pyplot as plt
import tensorflow as tf
import librosa
import keras.backend as K
from datetime import datetime
from sklearn.utils import class_weight
from tensorflow.keras.models import load_model
```

```python
# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess the data
x_train = np.expand_dims(x_train, axis=-1) / 255.0
x_test = np.expand_dims(x_test, axis=-1) / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

```python
x_train.shape
```

```python

# Build the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
model.fit(x_train, y_train, batch_size=256, epochs=5, validation_split=0.1, shuffle = True)
```

```python
res0 = model.evaluate(x_test, y_test)
```

```python
def makeTask():
    num_tasks = 20
    tasks = [[] for _ in range(num_tasks)]
    labels = np.argmax(y_train_new, axis = 1)
    x, y = [[] for _ in notToBeRemoved], [[] for _ in notToBeRemoved]
    for a, b in zip(x_test_new, y_test_new):
        x[notToBeRemoved.index(np.argmax(b))].append(a)
        y[notToBeRemoved.index(np.argmax(b))].append(b)
        
    x = [np.array(i) for i in x]
    y = [np.array(i) for i in y]
    
    for task in range(num_tasks):
        tasks[task].append(np.concatenate([x[i][task*3:(task+1)*3] for i in range(len(x))]))
        tasks[task].append(np.concatenate([y[i][task*3:(task+1)*3] for i in range(len(y))]))
        tasks[task].append(np.concatenate([x[i][num_tasks*3+task*2:num_tasks*3+(task+1)*2] for i in range(len(x))]))
        tasks[task].append(np.concatenate([y[i][num_tasks*3+task*2:num_tasks*3+(task+1)*2]for i in range(len(y))]))
    return tasks
```

```python
task_list = makeTask()
```

```python
loss_fn=tf.keras.losses.CategoricalCrossentropy()
precision_metric = keras.metrics.Precision()
recall_metric = keras.metrics.Recall()
optim = tf.keras.optimizers.legacy.Adam(
    learning_rate=0.0004
)

for epoch in range(20):
    print(f"start of epoch : {epoch}")
    
    
    with tf.GradientTape(persistent = True) as tape:
        
        for episode in range(5):
            outer_loss = None
            for task in task_list[episode*4:(episode+1)*4]:

                clone_model = tf.keras.models.clone_model(model)
                clone_model.set_weights(model.get_weights())



                y_pred = clone_model(task[0], training = True)
                loss = loss_fn(task[1], y_pred)
                gradients = tape.gradient(loss, clone_model.trainable_weights)
                optim.apply_gradients(zip(gradients, clone_model.trainable_weights))



                y_pred_new = clone_model(task[2], training = True)
                loss_new = loss_fn(task[3], y_pred_new)


                if outer_loss == None:
                    outer_loss = loss_new
                else:
                    outer_loss += loss_new

            clone_model.set_weights(model.get_weights())

            newGradients = tape.gradient(outer_loss, clone_model.trainable_weights)
            optim.apply_gradients(zip(newGradients, model.trainable_weights))
```
